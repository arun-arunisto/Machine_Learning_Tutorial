##Bar Charts and Heat Maps
- for this tutorial we are going to use US Department of Transportation flight delays dataset

```python
#loading the dataset
flight_data = pd.read_csv("/content/drive/MyDrive/Datascience&MachineLearning/datasets/flight_delays.csv", index_col="Month")
flight_data.head()
```

- Each entry shows the flight delays in every month

```python
#next we are going to generate a bar map using seaborn
plt.figure(figsize=(10, 6))

plt.title("Average Flight Delays")

#bar chart using seaborn
sns.barplot(x=flight_data.index, y=flight_data['NK'])

#adding label for y axis using plt
plt.ylabel("Arrival Delay in minutes")
```

##Heatmap using seaborn


```python
plt.figure(figsize=(14, 7))
plt.title("Average arrival delay for each airline")

#heatmap using seaborn
sns.heatmap(data=flight_data, annot=True)

plt.xlabel('Airline')
```

##Scatter Plots
- we are going to use synthetic dataset of insurance changes

```python
#loading the dataset
insurance_data = pd.read_csv('/content/drive/MyDrive/Datascience&MachineLearning/datasets/insurance.csv')
insurance_data.head()
```

```python
plt.figure(figsize=(14, 6))
#scatter plots using seaborn
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

- we can use the scatter plots to display the relationships between (not two, but) thre variables! one-way of doing is by color-coding the points
- for instance to understand how smoking affects the relationship between BMI and insurance costs, we can color-code the points by 'smoker', and plotting the other two columns ['bmi', 'charges']


```python
#color-coded scatter plots
plt.figure(figsize=(14, 6))

plt.title('Color-coded Scatter plots')

sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

###lmplot
- sns.lmplot add two regression lines, corresponding to given data

```python
# Create lmplot
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

###Swarmplot
- sns.swarmplot using to design plots for categorical variable

```python
#swarmplot
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

##Histograms and Density Plots
- for this we are going to work with a dataset of 150 different flowers, or 50 each from three different species of iris
<img src="https://storage.googleapis.com/kaggle-media/learn/images/RcxYYBA.png">

```python
#loading the dataset
iris_data = pd.read_csv('/content/drive/MyDrive/Datascience&MachineLearning/datasets/iris.csv', index_col='Id')
iris_data.head()
```

```python
plt.figure(figsize=(14, 6))

plt.title("Petal Length (in cm)")
#histograms using seaborn
sns.histplot(iris_data['Petal Length (cm)'])
```

- Density plots
the next type of plot is a "kernel density estimate (KDE)" plot.

```python
plt.figure(figsize=(14, 6))
plt.title("Petal Length in cm")
#kde plot using seaborn
sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True)
```

- and also we are able to create "two dimensional (2D) KDE plot" with sns.jointplot command

```python
#2D KDE plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```

- Color-coded plots
we are going to create plots to understand the differences between the species
for that we are going to use color coded plots

```python
#color coded histograms
plt.figure(figsize=(14, 6))
plt.title("Histogram of each species")

sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')
```

```python
#color-coded for kde plots
plt.figure(figsize=(14, 6))
plt.title("Distribution of petal lengths, by species")

sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', fill=True)
```

##Choosing Plot types and custom styles
1. for "Trends" - A trend is defined as a pattern of change
- for that "LIne charts"(sns.lineplot) are best to show
2. "Relationship" - There are many different chart types that you can use to understand relationships between variables in your data
- sns.barplot - Bar charts is useful for comparing
- sns.heatmap - Heatmaps can be used to find the color coded patterns in tables of numbers
- sns.scatterplot - Scatter plots show the relationship between two continuos variables, if color-coded we can also show the relationship with a third categorical variables
- sns.regplot - regression line in the scatter plot makes it easier to see any linear relationship between two variables
- sns.lmplot - command is useful for drawing multiple regression lines
- sns.swarmplot - categorical scatter plots show the relationship between contionus variable and ctegorical variable
3. Distribution - to show the possible values that we can expect to see in a variable along with how likely they are
- sns.histplot - Histograms show the distribution of single numerical variable
- sns.kdeplot - KDE plots (or 2D KDE plots) shows as estimated smooth distribution of a single or two numerical variables.
- sns.jointplot - this command is useful for simultaneously displaying a 2D KDE plot
